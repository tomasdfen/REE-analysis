````python
import tensorflow.keras.backend as K
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
from modules.read_data import cleanup
from tensorflow.keras.layers import Dense, Input,Layer
from tensorflow.keras.models import Model
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import scale
from typing import List
````

````python
class Autoencoder():
    def __init__(self, input_dim , encoder_units,encoder_acts, decoder_units, decoder_acts, n_clusters):
        
        assert len(encoder_units) == len(encoder_acts), "The amount of encoder layers doesn't fit the amount of its activations functions"
        assert len(decoder_units) == len(decoder_acts), "The amount of decoder layers doesn't fit the amount of its activations functions"
        
        self.name = 'autoencoder'
        
        self.input_dim = input_dim
        self.encoder_units = encoder_units
        self.decoder_units = decoder_units
        self.encoder_acts = encoder_acts
        self.decoder_acts = decoder_acts
        
        self._build()
        
    def _build(self):
        
        encoder_input = Input(shape = self.input_dim, name='encoder_input')
        x = encoder_input
        
        for i in range(len(encoder_units)):
            dense_layer = Dense(encoder_units[i], encoder_acts[i])
            x = dense_layer(x)
        
        encoder_output = Dense(self.n_clusters, name='encoder_output')(x)
        self.encoder = Model(encoder_input, encoder_output)
        
        decoder_input = Input(shape=(self.n_clusters,), name='decoder_input')
        
        x = Dense(self.encoder_units[-1])(x)
        
        for i in range(len(decoder_units)):
            dense_layer = Dense(decoder_units[i], decoder_acts[i])
            x = dense_layer(x)
        
        decoder_output = x
        
        self.decoder = Model(decoder_input, decoder_output)
        
        self.model = Model(encoder_input,self.decoder(encoder_output))
        
    def compile(optimizer, loss=None, metrics=None, loss_weights=None, sample_weight_mode=None, weighted_metrics=None, target_tensors=None):
        self.model.compile(optimizer, loss, metrics, loss_weights, sample_weight_mode, weighted_metrics, target_tensors)

    def load_weights(self, filepath):
        self.model.load_weights(filepath)
    
    def fit(x=None, y=None, batch_size=None, epochs=1, verbose=1, callbacks=None, validation_split=0.0, validation_data=None, shuffle=True, class_weight=None, sample_weight=None, initial_epoch=0, steps_per_epoch=None, validation_steps=None, validation_freq=1, max_queue_size=10, workers=1, use_multiprocessing=False):
        hist = self.model.fit(x, y, batch_size, epochs, verbose, callbacks, validation_split, validation_data, shuffle, class_weight, sample_weight, initial_epoch, steps_per_epoch, validation_steps, validation_freq, max_queue_size, worker, use_multiprocessing)
        return hist
````

````python
class ClusteringLayer(Layer):
    """
    Clustering layer converts input sample (feature) to soft label.

    # Example
    ```
        model.add(ClusteringLayer(n_clusters=10))
    ```
    # Arguments
        n_clusters: number of clusters.
        weights: list of Numpy array with shape `(n_clusters, n_features)` witch represents the initial cluster centers.
        alpha: degrees of freedom parameter in Student's t-distribution. Default to 1.0.
    # Input shape
        2D tensor with shape: `(n_samples, n_features)`.
    # Output shape
        2D tensor with shape: `(n_samples, n_clusters)`.
    """

    def __init__(self, n_clusters, weights=None, alpha=1.0, **kwargs):
        if 'input_shape' not in kwargs and 'input_dim' in kwargs:
            kwargs['input_shape'] = (kwargs.pop('input_dim'),)
        super(ClusteringLayer, self).__init__(**kwargs)
        self.n_clusters = n_clusters
        self.alpha = alpha
        self.initial_weights = weights
        self.input_spec = InputSpec(ndim=2)

    def build(self, input_shape):
        assert len(input_shape) == 2
        input_dim = input_shape[1]
        self.input_spec = InputSpec(dtype=K.floatx(), shape=(None, input_dim))
        self.clusters = self.add_weight((self.n_clusters, input_dim), initializer='glorot_uniform', name='clusters')
        if self.initial_weights is not None:
            self.set_weights(self.initial_weights)
            del self.initial_weights
        self.built = True

    def call(self, inputs, **kwargs):
        """ student t-distribution, as same as used in t-SNE algorithm.        
                 q_ij = 1/(1+dist(x_i, µ_j)^2), then normalize it.
                 q_ij can be interpreted as the probability of assigning sample i to cluster j.
                 (i.e., a soft assignment)
        Arguments:
            inputs: the variable containing data, shape=(n_samples, n_features)
        Return:
            q: student's t-distribution, or soft labels for each sample. shape=(n_samples, n_clusters)
        """
        q = 1.0 / (1.0 + (K.sum(K.square(K.expand_dims(inputs, axis=1) - self.clusters), axis=2) / self.alpha))
        q **= (self.alpha + 1.0) / 2.0
        q = K.transpose(K.transpose(q) / K.sum(q, axis=1)) # Make sure each sample's 10 values add up to 1.
        return q

    def compute_output_shape(self, input_shape):
        assert input_shape and len(input_shape) == 2
        return input_shape[0], self.n_clusters

    def get_config(self):
        config = {'n_clusters': self.n_clusters}
        base_config = super(ClusteringLayer, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))
````

````python
file = "datasets/demanda/datos-demanda-{m}-{y}.json"
data = pd.DataFrame()
dates = []
for year in (2017,2018,2019):
    for m in range(1,13):
        values = cleanup(file.format(m=m, y=year), 'minute')
        dates.append(f"{datetime(year, m, d).weekday}, {d}-{m}-{year}" for d in range(1,len(values)))
        data = data.append(values)
dates = pd.Series(dates)
````

````python
n_clusters = 20
models = [KMeans(n_clusters=i) for i in range(2,n_clusters+1)]
scores = [model.fit(data).score(data) for model in models]
sillhouette_scores = [silhouette_score(data, model.predict(data)) for model in models]
````
